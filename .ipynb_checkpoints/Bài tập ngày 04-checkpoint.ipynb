```python
import turtle
t = turtle.Turtle()
t.color("red","blue")
t.pensize(3)
t.penup()
t.goto(-100,100)
t.pendown()

t.begin_fill()
for i in range(2):
    t.fd(300)
    t.rt(90)
    t.fd(150)
    t.rt(90)
t.end_fill()
```

```python
import turtle
t = turtle.Turtle()
t.pensize(2)
turtle.bgcolor("lightgreen")


## vẽ mặt trời
t.speed(0)
t.penup()
t.goto(150,200)
t.pendown()
t.fillcolor("yellow")
t.begin_fill()
for i in range(4):
    t.fd(30)
    t.bk(30)
    t.lt(90)
    for i in range(45):
        t.fd(2*3.14*30/360)
        t.lt(1)
    t.rt(90)
    t.fd(15)
    t.bk(15)
    t.lt(90)
    for i in range(45):
        t.fd(2*3.14*30/360)
        t.lt(1)
    t.rt(90)
t.end_fill()

## vẽ nhà
t.penup()
t.goto(-200,0)
t.pendown()

t.fillcolor("pink")
t.begin_fill()
t.fd(88)
t.lt(150)
t.fd(50)
t.lt(60)
t.fd(50)
t.lt(60)
t.end_fill()

t.fillcolor("blue")
t.begin_fill()
for i in range(2):
    t.fd(120)
    t.lt(90)
    t.fd(88)
    t.lt(90)
t.end_fill()   

t.lt(90)
t.fd(88)
t.rt(90)
t.fillcolor("yellow")
t.begin_fill()
for i in range(2):
    t.fd(120)
    t.lt(110)
    t.fd(80)
    t.lt(70)
t.end_fill()

t.fillcolor("orange")
t.begin_fill()
t.lt(110)
t.fd(80)
t.lt(130)
t.fd(50)
t.lt(50)
t.fd(80)
t.end_fill()

t.penup()
t.goto(-200,0)
t.setheading(-90)
t.fd(120)
t.lt(90)
t.fd(20)
t.pendown()
t.fillcolor("green")
t.begin_fill()
for i in range(2):
    t.fd(40)
    t.lt(90)
    t.fd(60)
    t.lt(90)
t.end_fill()

t.penup()
t.goto(-200,0)
t.setheading(0)
t.fd(80)
t.rt(35)
t.fd(40)
t.setheading(-90)
t.pendown()

t.fillcolor("brown")
t.begin_fill()
for i in range(2):
    t.fd(30)
    t.lt(110)
    t.fd(30)
    t.lt(70)
t.end_fill()
   
## vẽ cây thông

t.setheading(0)
t.penup()
t.goto(200,-200)
t.pendown()

t.fillcolor("brown")
t.begin_fill()
for i in range(2):
    t.fd(16)
    t.lt(90)
    t.fd(35)
    t.lt(90)
t.end_fill()

t.penup()
t.goto(208,-200)
t.lt(90)
t.fd(35)
t.rt(90)
t.pendown()


for i in range(3):
    t.bk(30)
    t.fillcolor("green")
    t.begin_fill()
    for i in range(3):
        t.fd(60)
        t.lt(120)
    t.end_fill()
    t.penup()
    t.lt(60)
    t.fd(60)
    t.pendown()
    t.rt(60)
    


```
